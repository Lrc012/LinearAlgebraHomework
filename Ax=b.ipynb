```python
class Gauss:
    def __init__(self):
        print("解矩阵方程Ax=b")
        print("输入A矩阵的行/列n：")
        self.n = int(input())
        print("依次输入A矩阵的每一行，一行以空格隔开每个数")
        self.A = [[0]*self.n]*self.n
        for i in range(self.n):
            self.A[i] = input().split(" ")
            for j in range(self.n):
                self.A[i][j] = int(self.A[i][j])
        print("输入b矩阵的列数m：")
        self.m=int(input())
        print("依次输入b矩阵的每一行，一行以空格隔开每个数")
        self.b = [[0]*self.m]*self.n
        for i in range(self.n):
            self.b[i] = input().split(" ")
            for j in range(self.m):
                self.b[i][j] = int(self.b[i][j])
```

```python
import numpy as np
opr=Gauss()
print(np.linalg.solve(np.array(opr.A),np.array(opr.b)))
```
