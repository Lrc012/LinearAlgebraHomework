```python
class Gauss:
    def __init__(self):
        print("输入矩阵的行/列数：")
        self.n = int(input())
        print("依次输入矩阵的每一行，一行以空格隔开每个数")
        self.matrix = [[0]*self.n]*self.n
        for i in range(self.n):
            self.matrix[i] = input().split(" ")
            for j in range(self.n):
                self.matrix[i][j] = int(self.matrix[i][j])
```

```python
import numpy as np
opr=Gauss()
print(np.matrix(np.array(opr.matrix)).I)
```
